```python
from packing_algorithm import Packer
from bin import Bin
from box import Box
import matplotlib.pyplot as plt
import pandas as pd
import time
from sklearn.utils import shuffle
import tracemalloc
# allowed_rots = [0, 1, 2, 3, 4, 5]

packer = Packer()

def generate_boxes(shuffle = False, nb_of_S = 0, nb_of_M = 0, nb_of_L = 0, nb_of_XL = 0, nb_of_T8=0, nb_of_T10=0):
    index = 1
    for i in range(nb_of_S):
        packer.add_item(Box(f'S [box {index}]', 20, 11, 11))
        index += 1
    for i in range(nb_of_M):
        packer.add_item(Box(f'M [box {index}]', 30, 20, 25))
        index += 1
    for i in range(nb_of_L):
        packer.add_item(Box(f'L [box {index}]', 53, 23, 32))
        index += 1
    for i in range(nb_of_XL):
        packer.add_item(Box(f'XL [box {index}]', 70, 32, 40))
        index += 1
    for i in range(nb_of_T8):
        packer.add_item(Box(f'T8 [box {index}]', 40, 6, 29))
        index += 1
    for i in range(nb_of_T10):
        packer.add_item(Box(f'T10 [box {index}]', 30, 60, 30))
        index += 1
    if shuffle==True:
        shuffle_boxes()
        
def shuffle_boxes():
    packer.items_to_pack=shuffle(packer.items_to_pack)

def generate_bins(nb_of_bins = 10, width = 117, height = 120, depth = 295):
    #default bin is 2 times smaller than OG 
    for i in range(nb_of_bins):
        packer.add_bin(Bin(f'20-standard [bin {i}]', width, height, depth))

```

```python
def test_diff_size_container(bins_width = 117, bins_height = 120, bins_depth = 295, nb_of_size_step = 100, size_step = 0.1):
    
    memory_package_array= []
    time_to_process_array= []
    index_array=[]

    for i in range(nb_of_size_step):
        packer = Packer()
        generate_boxes(shuffle = False,nb_of_S = 50, nb_of_M = 50, nb_of_L = 50, nb_of_XL = 50, nb_of_T8=50, nb_of_T10=50)
        width = bins_width + bins_width * i * size_step
        height = bins_height + bins_height * i * size_step
        depth = bins_depth + bins_depth * i * size_step
        generate_bins(nb_of_bins = 10, width = width, height = height, depth = depth)

        tracemalloc.start()
        start = time.time()
        packer.pack(distribute_items=True,bigger_first= True)
        end = time.time()
        memory_usage = tracemalloc.get_traced_memory()
        tracemalloc.stop()
        memory_package_array.append(memory_usage[1])
        time_to_process_array.append(end-start)
        index_array.append(1+i*size_step)


    return memory_package_array, time_to_process_array, index_array
```
